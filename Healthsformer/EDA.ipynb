```python
res = pd.read_csv("queryResults0.5.csv")
res = res.drop_duplicates()
res[['admission_time', 'ED_intime', 'ED_outtime', 'discharge_time', 'TPA_time', 'CT_time', 'ICU_intime', 'ICU_outtime', 'SLP_time', 'SW_time']] = res[['admission_time', 'ED_intime', 'ED_outtime', 'discharge_time', 'TPA_time', 'CT_time', 'ICU_intime', 'ICU_outtime', 'SLP_time', 'SW_time']].astype('datetime64[ns]')
res['stay_id'] = res['stay_id'].astype(pd.Int64Dtype())
res = res[['subject_id', 'hadm_id', 'stay_id', 'icd_code', 'gender', 'anchor_age',
           'admission_time', 'discharge_time', 'ED_intime', 'ED_outtime', 'TPA_time',
           'CT_time', 'ICU_intime', 'ICU_outtime', 'ICU_los', 'SLP_time', 'SW_time']]
print(res.shape)
```

```python
selection = 'sub'
if selection == 'all':
    # all icd codes
    data = res
elif selection == 'all+':
    # all icd codes of length 7+
    data = res[res.isna().sum(axis=1) < 3]
elif selection == 'sub':
    # blood clot icd codes
    data = res[res['icd_code'].isin(codes)]
else:
    # blood clot icd codes of length 7+
    data = res[res['icd_code'].isin(codes)]
    data = data[data.isna().sum(axis=1) < 3]

indices = (data.groupby(['subject_id', 'hadm_id', 'stay_id']).nunique()['ED_intime']>1).reset_index()
asd = indices[indices['ED_intime']]
data = data[~data['subject_id'].isin(asd['subject_id'].values)]
data
```

```python
keys = (data[['subject_id', 'hadm_id']].groupby(['subject_id', 'hadm_id']).size()).reset_index()[['subject_id', 'hadm_id']]
sequences = []

singular_columns = ['subject_id', 'hadm_id', 'stay_id', 'icd_code', 'gender', 'anchor_age', 'admission_time',
                    'discharge_time', 'ED_intime', 'ED_outtime']
duplicatables = ['TPA_time', 'CT_time', 'SLP_time', 'SW_time']
stay_id_dependent_columns = ['ICU_intime', 'ICU_outtime', 'ICU_los']


for rownum in range(len(keys)):
    sequence = {}
    sub = data[(data['subject_id']==keys.iloc[rownum, 0]) & (data['hadm_id']==keys.iloc[rownum, 1])]
    for col in singular_columns:
        if not sub[col].isna().any():
            sequence[col] = sub[col].unique()[0]
            

    for col in duplicatables:
        i = 0
        uniques = sub[col].dropna().unique()
        if len(uniques) > 1:
            for var in uniques:
                sequence[f'{col}{i}'] = sub[sub[col]==var][col].values[0]
                i += 1
            
    i = 0
    for stayid in sub['stay_id'].dropna().unique():
        if len((sub[sub['stay_id']==stayid])[['ICU_intime', 'ICU_outtime', 'ICU_los']]) != 0:
            sequence[f'icuin{i}'] = sub[sub['stay_id']==stayid]['ICU_intime'].drop_duplicates().values[0]
            sequence[f'icuout{i}'] = sub[sub['stay_id']==stayid]['ICU_outtime'].drop_duplicates().values[0]
            sequence[f'iculos{i}'] = sub[sub['stay_id']==stayid]['ICU_los'].drop_duplicates().values[0]
        i += 1
    sequences.append(sequence)
```

```python
uniq = (data.groupby(['subject_id', 'hadm_id']).nunique()).reset_index()
avg_length = uniq[['admission_time', 'discharge_time', 'ED_intime', 'ED_outtime', 'TPA_time', 'CT_time', 'SLP_time', 'SW_time']].sum(axis=1).mean()
print(len(sequences))
print(avg_length)
```

# Sequence metadata
- A patient who is admitted to the ICU is issued a stay_id 
- for each stay_id, social workers can visit many times (like 15 even)
- for some combinations of subject_id and hadm_id, there exist duplicate events, such as more than 1 ct scan, icu entry, and so on.
- for every combination of subject_id and hadm_id we have 1 sequence of events.
    - for every sequence of events, we can have multiple sub-sequences of events based on stay_id which consists of ct_time, icu_intime, icu_outtime, icu_los
- icu values in the sequences are labeled as icu0, icu1, ...
- all type of icu are considered
